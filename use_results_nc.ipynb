```python
import glob
import xarray as xr
import numpy as np
```

```python
# create an empty xarray to fill with results
original_data = xr.open_dataset('/climca/data/CESM2_LE/TREFHTMN/mon_raw/b.e21.BHISTcmip6.f09_g17.LE2-1001.001.cam.h0.TREFHTMN.201001-201412.nc')['TREFHTMN']
original_data = original_data.loc[:,10:20,10:20]

results1 = original_data.copy() * np.nan
```

```python
for file_name in glob.glob("output/nc/result1_*.nc"):
    with xr.load_dataset(file_name) as nc:
        nearest = results1.sel(dict(lon=nc.lon, lat=nc.lat), method='nearest')
        results1.loc[:,nearest.lat,nearest.lon] = nc['result1']
```
